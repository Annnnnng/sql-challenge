```python
# import Pandas, SQLalchemy and psycopg2
import pandas as pd
import matplotlib.pyplot as plt
from sqlalchemy import create_engine
import psycopg2
```

```python
# Create engine
engine = create_engine('postgresql://postgres:La871121@localhost:5432/homework')
```

```python
# Create connection
conn = engine.connect()
```

```python
# Display employee salary in the database
salary = pd.read_sql("SELECT * FROM salaries", conn)
salary.head()
```

```python
#Create a histogram to visualize the most common salary ranges for employees
plt.hist(salary["salary"])
plt.show()
```

```python
# Query all employee record in the database
emp_data = pd.read_sql("SELECT * FROM employees", conn)
```

```python
# Display employee data
emp_data.head()
```

```python
# Query employee title record in the database
title = pd.read_sql("SELECT * FROM titles", conn)
title.head()
```

```python
# Rename column title_id to emp_title_id to make the column name consistant 
titles = title.rename(columns={"title_id": "emp_title_id"})
titles
```

```python
# Merge employee titles with employee records by employee title id
emp_title = pd.merge(emp_data, titles, on = "emp_title_id", how="inner")
emp_title
```

```python
# Merge employee salary record with employee records 
emp_salary = pd.merge(emp_title, salary, on = "emp_no")
emp_salary
```

```python
# Group employee record by title and calculate average salary
ave_salary = emp_salary.groupby(["title"]).mean()
ave_salary_title = ave_salary.reset_index()
ave_salary_title
```

```python
# Plot bar chart to show average salary by title
x = ave_salary_title["title"]
y = ave_salary_title["salary"]
plt.bar(x,y)
plt.xlabel("Title", fontsize = 20)
plt.ylabel("Average Salary", fontsize = 20)
plt.xticks(rotation=90)
plt.show()
```
